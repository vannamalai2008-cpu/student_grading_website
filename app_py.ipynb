```python
# 1. Install Streamlit and Ngrok helper
!pip install -q streamlit pyngrok

# 2. Write your student grading code to app.py
with open("app.py", "w") as f:
    f.write('''
import streamlit as st

st.set_page_config(page_title="Student Grading Dashboard", page_icon="📝", layout="centered")

st.title("👨‍🎓 Student Grading Web Dashboard")
st.write("Enter the marks for each subject (Out of 100) to check real-time results.")

col1, col2 = st.columns(2)

with col1:
    math = st.number_input("Mathematics", min_value=0, max_value=100, value=75)
    science = st.number_input("Science", min_value=0, max_value=100, value=65)
    english = st.number_input("English", min_value=0, max_value=100, value=38)

with col2:
    history = st.number_input("History", min_value=0, max_value=100, value=82)
    computer = st.number_input("Computer Science", min_value=0, max_value=100, value=90)

subjects = {"Mathematics": math, "Science": science, "English": english, "History": history, "Computer Science": computer}

st.markdown("---")
st.subheader("📊 Subject-wise Report Card")

all_passed = True
total_marks = 0

for subject, mark in subjects.items():
    total_marks += mark
    if mark >= 40:
        st.success(f"✅ **{subject}**: {mark}/100 — **PASS**")
    else:
        st.error(f"❌ **{subject}**: {mark}/100 — **FAIL**")
        all_passed = False

st.markdown("---")
st.subheader("📝 Summary")

percentage = (total_marks / 500) * 100
st.metric(label="Overall Percentage", value=f"{percentage:.2f}%")

if all_passed:
    st.success("🎉 FINAL RESULT: PASSED THE SEMESTER!")
else:
    st.error("⚠️ FINAL RESULT: FAILED (Must clear all individual subjects)")
''')

# 3. Authenticate with Ngrok (REPLACE WITH YOUR ACTUAL AUTHTOKEN FROM NGROK.COM)
from pyngrok import ngrok
ngrok.set_auth_token("3GXFGaniSu1GSLKIs08PFLHDNZp_4rgnbBTKVE6nVNw4RWGYi")

# 4. Open the public web link tunnel
public_url = ngrok.connect(8501, proto="http")
print("\n" + "="*60)
print("👨‍💻 YOUR STABLE PUBLIC WEBSITE LINK IS READY!")
print(f"🔗 Click here: {public_url}")
print("="*60 + "\n")

# 5. Boot up the Streamlit engine
!streamlit run app.py --server.port 8501 --server.headless true
```
